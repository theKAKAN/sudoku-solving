```python
import re
import matplotlib.pyplot as plt

# Specify the file path
file_path = "test.txt"  # Replace with the actual path

# Initialize empty dictionaries for storing data
data = {}
current_epoch = None

with open(file_path, "r") as file:
    for line in file:
        # Check for epoch line
        match_epoch = re.search(r"Epoch (\d+)", line)
        if match_epoch:
            current_epoch = int(match_epoch.group(1))
            data[current_epoch] = {}  # Create a dictionary for this epoch

        # Check for accuracy line (assuming it starts with "accuracy:")
        match_accuracy = re.search(r"accuracy: (\d+\.\d+)", line)
        if match_accuracy and current_epoch is not None:
            data[current_epoch]["accuracy"] = float(match_accuracy.group(1))

# Extract epochs and accuracies
epochs = []
accuracies = []
for epoch, values in data.items():
    if "accuracy" in values:
        epochs.append(epoch)
        accuracies.append(values["accuracy"])

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(epochs, accuracies, label="Accuracy")

# Add labels and title
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Epoch")

# Customize the plot (optional)
plt.grid(True)
plt.xticks(epochs)

# Display the plot
plt.show()
```
